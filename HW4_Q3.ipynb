```python
import torch
import torch.nn.functional as F
import math

def scaled_dot_product_attention(Q, K, V):
    """
    Q: (batch, heads, seq_len_q, d_k)
    K: (batch, heads, seq_len_k, d_k)
    V: (batch, heads, seq_len_k, d_v)
    """
    d_k = Q.size(-1)

    # raw scores
    scores = torch.matmul(Q, K.transpose(-2, -1))  # (batch, heads, seq_len_q, seq_len_k)
    print("Raw scores (before scaling):")
    print(scores)

    # scaled scores
    scores_scaled = scores / math.sqrt(d_k)
    print("\nScaled scores (after dividing by sqrt(d_k)):")
    print(scores_scaled)

    # softmax over keys
    attn_weights = F.softmax(scores_scaled, dim=-1)
    print("\nAttention weights (after softmax):")
    print(attn_weights)

    # output
    output = torch.matmul(attn_weights, V)
    print("\nOutput vectors (Attention(Q,K,V)):")
    print(output)

    return output, attn_weights


if __name__ == "__main__":
    torch.manual_seed(0)

    batch_size = 1
    num_heads = 1
    seq_len_q = 3
    seq_len_k = 3
    d_k = 4
    d_v = 4

    Q = torch.randn(batch_size, num_heads, seq_len_q, d_k)
    K = torch.randn(batch_size, num_heads, seq_len_k, d_k)
    V = torch.randn(batch_size, num_heads, seq_len_k, d_v)

    print("Q:\n", Q)
    print("\nK:\n", K)
    print("\nV:\n", V)

    output, attn_weights = scaled_dot_product_attention(Q, K, V)
```
